```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Sample microbiome data (replace this with your actual data)
microbiome_data = pd.DataFrame({
    'Sample': ['Sample1', 'Sample2', 'Sample3'],
    'Taxon1': [100, 150, 200],
    'Taxon2': [50, 80, 120],
    'Taxon3': [30, 40, 50]
})

# Set the 'Sample' column as the index (if it's not already)
microbiome_data.set_index('Sample', inplace=True)

# Bar plot
microbiome_data.plot(kind='bar', stacked=True)
plt.title('Taxonomic Composition Across Samples')
plt.xlabel('Samples')
plt.ylabel('Abundance')
plt.show()

# Stacked bar plot
microbiome_data.plot(kind='bar', stacked=True)
plt.title('Taxonomic Composition Across Samples (Stacked)')
plt.xlabel('Samples')
plt.ylabel('Abundance')
plt.show()

# Alpha diversity plot (e.g., Shannon diversity index)
shannon_diversity = microbiome_data.sum(axis=1).plot(kind='bar')
plt.title('Alpha Diversity Across Samples (Shannon Index)')
plt.xlabel('Samples')
plt.ylabel('Shannon Index')
plt.show()

# Differential abundance analysis (e.g., volcano plot)
# Replace the following code with your actual differential abundance analysis code
# Here, we'll just plot random data
import numpy as np

random_data = pd.DataFrame({
    'Taxon': ['Taxon1', 'Taxon2', 'Taxon3'],
    'Fold Change': np.random.uniform(low=-2, high=2, size=3),
    'P-value': np.random.uniform(low=0, high=0.05, size=3)
})

plt.scatter(random_data['Fold Change'], -np.log10(random_data['P-value']))
plt.axhline(y=-np.log10(0.05), color='r', linestyle='--')
plt.xlabel('Fold Change')
plt.ylabel('-log10(P-value)')
plt.title('Volcano Plot for Differential Abundance Analysis')
plt.show()

# Summary statistics
summary_stats = microbiome_data.describe()
print(summary_stats)
```
